```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# transform data
df_scaled =scaler.fit_transform(df_cleaned)
df_scaled= pd.DataFrame(df_scaled,columns=df_cleaned.columns)
df_scaled
```

```python
# training and testing data

from sklearn.model_selection import train_test_split

x = df_scaled.drop(["HeartDisease"],axis=1)
y = df_scaled['HeartDisease']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
# sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))

# derivative of sigmoid function
def sigmoid_p(x):
    return sigmoid(x) * (1-sigmoid(x))    
```

```python
#weight and bais
weight=np.array([])
for i in range(14):
    w=np.random.randn()
    weight=np.concatenate((weight,np.array([w])))


weight=weight.reshape(14,1)
b=np.random.randn()

weight=np.matrix(weight)


# training loop
 
X_train=np.matrix(X_train)
y_train=y_train.reshape(642,1)
y_train=np.matrix(y_train)

weight=np.matrix(weight)

learning_rate = 0.01

errors = []

for i in range(13000): 
    ri = np.random.randint(len(X_train))  
    
    z = (X_train[ri] @ weight)[0,0] + b
    pred = sigmoid(z)   
    

    target = y_train[ri] 


    # error for current random point
    error = np.square(pred-target[0,0])
    

    # print the cost over all data points every 1k iters
    if i % 10 == 0:
       c = 0
       n=len(X_train)
       for j in range(n):
           p = X_train[j]
           p_pred = sigmoid( (p @ weight)[0,0] + b)
           c += np.square(p_pred - y_train[j][0,0])
       errors.append(c/n)
    
    




    derror_pred = 2 * (pred-target[0,0])
    dpred_dz = sigmoid_p(z)

    derror_dz = derror_pred * dpred_dz
    
    dz_dw = np.matrix.transpose(X_train[ri])
    dz_db = 1
    
    derror_dw = derror_dz * dz_dw
    derror_db = derror_dz * dz_db

    weight = weight - (learning_rate * derror_dw)
    b = b - (learning_rate * derror_db)
   

plt.plot(errors)   
```
